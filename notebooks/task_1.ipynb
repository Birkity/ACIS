```python
import csv
import pandas as pd
from pathlib import Path
```

```python
input_path = Path("../Data/MachineLearningRating_v3.txt")
output_path = Path("data.csv")


df = pd.read_csv(input_path, sep='|')


df.to_csv(output_path, index=False)


df.head()
```

Data Understanding 

```python
print(df.head())

print(df.info())

print(df.dtypes)
```

```python
print(df.describe())
```

Descriptive Statistics

```python
total_premium_stats = df['TotalPremium'].describe()
total_claims_stats = df['TotalClaims'].describe()
print(total_premium_stats)
print(total_claims_stats)
```

Data Quality Assessment

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df['TotalPremium'].hist(bins=50)
plt.title('Total Premium Distribution')
plt.show()

df['TotalClaims'].hist(bins=50)
plt.title('Total Claims Distribution')
plt.show()

df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()
```

```python
corr_matrix = df[['TotalPremium', 'TotalClaims']].corr()
print(corr_matrix)

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation between Total Premium and Total Claims')
plt.show()
```

```python
df.plot.scatter(x='TotalPremium', y='TotalClaims')
plt.title('Total Premium vs. Total Claims')
plt.show()
```

```python
sns.boxplot(x=df['TotalPremium'])
plt.title('Boxplot for Total Premium')
plt.show()

sns.boxplot(x=df['TotalClaims'])
plt.title('Boxplot for Total Claims')
plt.show()
```

```python
df.groupby('Province')['TotalPremium'].mean().plot(kind='bar')
plt.title('Average Premium per Province')
plt.show()

df.groupby('PostalCode')['TotalPremium'].mean().plot(kind='line')
plt.title('Average Premium by PostalCode')
plt.show()
```

```python
sns.boxplot(data=df, x='VehicleType', y='TotalClaims')
plt.title('Total Claims by Vehicle Type')
plt.show()
```
